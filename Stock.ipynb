```python
import yfinance as yf
```

```python
bse_sensex = yf.Ticker("^BSESN")
```

```python
bse_sensex = bse_sensex.history(period ="max")
```

```python
bse_sensex = bse_sensex.drop(columns = 'Dividends')
```

```python
bse_sensex = bse_sensex.drop(columns = 'Stock Splits')
```

```python
bse_sensex.head()
```

```python
bse_sensex["Tomorrow_Closing"] = bse_sensex["Close"].shift(-1)
```

```python
bse_sensex
```

```python
bse_sensex["Target"] = (bse_sensex["Tomorrow_Closing"] > bse_sensex["Close"]).astype(int)
```

```python
bse_sensex
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators =100 , min_samples_split =100 , random_state =1)
train = bse_sensex.iloc[:-100]
test = bse_sensex.iloc[-100:]
predictors =["Close" , "Open" , "High" , "Low", "Volume"]
model.fit(train[predictors], train["Target"])
```

```python
from sklearn.metrics import precision_score
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
preds = model.predict(test[predictors])
```

```python
preds = pd.Series(preds , index = test.index)
```

```python
preds
```

```python
precision_score(test['Target'], preds)
```

```python
combined = pd.concat ([test['Target'],preds] , axis=1)
```

```python
def predict(train ,test , predictors, model):
  model.fit(train[predictors], train["Target"])
  preds = model.predict(test[predictors])
  preds = pd.Series(preds , index=test.index , name = "Predictions")
  combined = pd.concat([test["Target"], preds], axis =1)
  return combined
```

```python
def backtest(data , model , predictors , start=2500 , step =250):
  all_predictions = []
  for i in range(start , data.shape[0], step):
    train = data.iloc[0:i].copy()
    test = data.iloc[i:(i+step)].copy()
    predictions = predict(train , test , predictors , model)
    all_predictions.append(predictions)
  return pd.concat(all_predictions)
```

```python
predictions = backtest(bse_sensex , model , predictors )
```

```python
precision_score(predictions["Target"], predictions["Predictions"])
```

```python
horizons = [2,5,30 ,60,250,500,1000]
new_predictors =[]
for horizon in horizons :
  rolling_averages = bse_sensex.rolling(horizon).mean()
  ratio_column = f"Close_Ratio_{horizon}"
  bse_sensex[ratio_column] = bse_sensex["Close"] / rolling_averages['Close']
  trend_column = f"Trend_{horizon}"
  bse_sensex[trend_column] =bse_sensex.shift(1).rolling(horizon).sum()['Target']

  new_predictors += [ratio_column , trend_column]
```

```python
bse_sensex = bse_sensex.dropna()
```

```python
model = RandomForestClassifier(n_estimators=200 , min_samples_split =50 , random_state =1)
```

```python
def predict(train ,test , predictors, model):
  model.fit(train[predictors], train["Target"])
  preds = model.predict_proba(test[predictors])[:,1]
  preds[preds >= 0.6] =1
  preds[preds < 0.6] =0
  preds = pd.Series(preds , index=test.index , name = "Predictions")
  combined = pd.concat([test["Target"], preds], axis =1)
  return combined
```

```python
predictions = backtest(bse_sensex , model , new_predictors)
```
